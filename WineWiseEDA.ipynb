#Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Loading Dataset

```python
dataset = pd.read_csv('winequality-white.csv', sep=';')
```

```python
print(dataset.head())
```

```python
print(dataset.isnull().sum())
print(dataset.describe())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=dataset)
plt.title('Distribution of white wine Quality Scores')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.show()
```

```python
dataset['quality_class'] = (dataset['quality']>6).astype(int)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='quality_class', data=dataset)
plt.title('Distribution of Binary White Wine Quality Classes')
plt.xlabel('Quality Class (0: Bad, 1: Good)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = dataset.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Features and Target')
plt.show()
```

```python
class_counts = dataset['quality_class'].value_counts()
class_proportions = dataset['quality_class'].value_counts(normalize=True) * 100
print('Class Counts:')
print(class_counts)
print('Class Proportions:')
print(class_proportions)
```
